```python
# import modules
import numpy as np
# astropy provides unit system for astronomical calculations
import astropy.units as u
# astropy also offer a module for showinng results in a table
import astropy.table as tbl
# import previous HW functions
from ReadFile import Read
from CenterOfMass import CenterOfMass
# astropy stores constants
from astropy.constants import G
# import modules for plotting
import matplotlib
import matplotlib.pyplot as plt
```

```python
class MassProfile:
    
    def __init__(self, galaxy, snap): # Initializing class
        
        # add a string of the filename to the value "000"
        ilbl = '000' + str(snap)
        # remove all but the last 3 digits 
        ilbl = ilbl[-3:]
        self.filename = "%s_"%(galaxy) + ilbl + '.txt'
        
        
        # read in data for x,y,z positions and mass
        self.time, self.total, self.data = Read(self.filename)
        
        # store x,y,z in appropriate units 
        self.m = self.data['m']
        self.x = self.data["x"]*u.kpc
        self.y = self.data["y"]*u.kpc
        self.z = self.data["z"]*u.kpc
        array_r = np.sqrt(self.x**2 + self.y**2 + self.z**2)
        
        # store the name of galaxy as a global property
        self.gname = galaxy
        
      
    def MassEnclosed(self, ptype, array_r):
        # Function to compute the mass enclosed within a desired radius of the COM position for a specified galaxy and its component
        # Inputs:
        #        ptype is the particle type of the specified galaxy
        #        array_r is the array with radii
        # Returns: 
        #        array_m is the array of masses (Msun)
        
        # calling COM_P function 
        COM = CenterOfMass(self.filename, ptype) # calling CenterOfMass class 
        COMP = COM.COM_P(0.1) # calling COM_P function within CenterOfMass class
        index1 = np.where(self.data['type'] == ptype) # index for desired Particle Type
        # Local Variables (don't need "self")
        x2 = self.x[index1] - COMP[0] # x-component of the enclosed radius
        y2 = self.y[index1] - COMP[1] # y-component of the enclosed radius
        z2 = self.z[index1] - COMP[2] # z-component of the enclosed radius
        r2 = np.sqrt(x2**2 + y2**2 + z2**2) # Radius of the enclosed radius
        m2 = self.m[index1]*1e10 # indexing Mass with desired Particle Type (multiplying 1e10 to make it units of 1e10 Msun)
        array_m = np.zeros(np.size(array_r)) # initializing an empty array 
        
        for i in range(np.size(array_r)): # looping over the defined array_r 
            
            index2 = np.where(r2<array_r[i]*u.kpc) # index for the condition that is the radius of all the particles must be less than the radius specified on the array 
            
            M = np.sum(m2[index2]) # summing all the mass of particles satisfying index2
            
            array_m[i] = M # populationg the empty array with all the particle mass per radius
                 
        return array_m*u.Msun # convert the mass array in Msun
   
    def MassEnclosedTotal(self, array_r):
        # Function to compute the Total Mass Enclosed 
        # Inputs:
        #        array_r is the array of radii
        # Returns: 
        #        array_M is the Total Mass (bulge+disk+halo) at radius each input array (Msun)
        
        array_M = [] # creating an empty array 
        
        if (self.gname == 'MW' or self.gname == 'M31'):
        
            for i in range(1,4): # for loop that loops i value from 1 to 3
            
                a = self.MassEnclosed(i, array_r) # running the MassEnclosed function with ptype = 1 to 3
                b = array_M.append(a) # populating the empty array with array of masses based on the array of radii
            
            c = np.concatenate((array_M[0],array_M[1],array_M[2]),axis=None) # combininng array of masses for ptype = 1,2,3
        
        else: # caveat for the absense of bulge in M33
            
            for i in range(1,3): # for loop that loops i value from 1 to 2
            
                a = self.MassEnclosed(i, array_r) # running the MassEnclosed function with ptype = 1 to 2
                b = array_M.append(a) # populating the empty array with array of masses based on the array of radii
            
            c = np.concatenate((array_M[0],array_M[1]),axis=None) # combininng array of masses for ptype = 1,2
    
                  
        return c

    
    def HernquistMass(self, array_r, a, Mhalo):
        # Function to compute the mass enclosed within a given radius using theoretical profile
        # Inputs: 
        #        array_r is the radius
        #        a is the scale factor
        #        Mhalo is the Halo Mass (Msun)
        # Returns:
        #       Mhalo is the Halo Mass (Msun)
        
        # Mass enclosed theoretical profile
        # M(r) = (Mhalo*r^2)/(a+r)^2 
        
        a = Mhalo*array_r**2
        b = (a+array_r)**2
        c = (a/b)/1e10 # converting to Msun units
        
        return c*u.Msun
        
    def CircularVelocity(self,ptype, array_r):
        # Inputs:
        #        ptype is the particle type of the specified galaxy
        #        array_r is the array with radii
        # Returns:
        #        array_Vc is the array of circular speeds (km/s)
        Gr = G.to(u.kpc*u.km**2/u.s**2/u.Msun) # Converting units of G to make Vcirc in units of km/s
        
        a = self.MassEnclosed(ptype, array_r) # calling MassEnclosed function from earlier to get the Mass Enclosed formula
        
        Vc =  np.sqrt(Gr*a/array_r) # Computing Circular Velocity = sqrt(G*M/r)
    
        return np.around(Vc,2) # rounding Circular Velocity into 2 decimal places
    
    def CircularVelocityTotal(self, array_r):
        # Inputs:
        #        array_r is the array of radii
        # Returns:
        #        The array of circular velocity (km/s) created by all galaxy components (bulge+disk+halo) at each array of radius
        Gr = G.to(u.kpc*u.km**2/u.s**2/u.Msun) # Converting units of G to make Vcirc in units of km/s
        
        array_TotalVc =[]
        
        for i in range(1,4):
            
            a = self.CircularVelocity(i, array_r) # calling MassEnclosed function from earlier to get the Mass Enclosed formula
            b = array_TotalVc.append(a)
            
        c = np.concatenate((array_TotalVc[0],array_TotalVc[1],array_TotalVc[2]),axis=None) 
    
        
        return c
    
    def HernquistVCirc(self, array_r, a, Mhalo):
        # Function that computes the circular speed using the Hernquist mass profile 
        # Inputs:
        #        array_r is the radius
        #        a is the scale factor 
        #        Mhalo is the halo mass
        # Returns:
        #        b is the Circular speed (km/s) 
        Gr = G.to(u.kpc*u.km**2/u.s**2/u.Msun) # Converting units of G to make Vcirc in units of km/s

        a = self.HernquistMass(array_r, a, Mhalo) # Calling Hernquist mass function
        
        b = np.sqrt(Gr*a/array_r) # Herquist Circular Velocity = sqrt(G*M/r)
        
        return b
    

       

    
    

        
        
        
    

    
```
